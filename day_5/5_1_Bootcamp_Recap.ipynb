````python
# Exercise 1: Variables and Basic Operations
num1 = 10
num2 = 5

# Perform operations
addition = num1 + num2
subtraction = num1 - num2
multiplication = num1 * num2
division = num1 / num2

# Print the results
print("Addition:", addition)
print("Subtraction:", subtraction)
print("Multiplication:", multiplication)
print("Division:", division)
````

````python
# Exercise 2: List Manipulation
star_trek_characters = ["Kirk", "Spock", "McCoy", "Scotty"]

# Append "Uhura" to the list
star_trek_characters.append("Uhura")

# Remove the second element
del star_trek_characters[1]

# Insert "Sulu" at the third position
star_trek_characters.insert(2, "Sulu")

# Sort the list in alphabetical order
star_trek_characters.sort()

# Print the modified list
print("Modified Star Trek Characters:", star_trek_characters)
````

````python
# Exercise 3: Looping and Conditionals
numbers = [10, 20, 30, 40, 50]

# Looping with for loop
print("Numbers:")
for number in numbers:
    print(number)

# Looping with while loop to find the first number divisible by 3
i = 0
while i < len(numbers):
    if numbers[i] % 3 == 0:
        print("First number divisible by 3:", numbers[i])
        break
    i += 1

# Conditional statement to check the length of the numbers list
if len(numbers) > 5:
    print("List is long enough")
else:
    print("List is too short")
````

````python
# Exercise 4: DataFrame Manipulation
import pandas as pd

# Student data
students_data = {
    "Name": ["John", "Emma", "Mike", "Emily"],
    "Age": [20, 22, 21, 19],
    "Grade": [85, 92, 88, 78]
}

# Create DataFrame
students_df = pd.DataFrame(students_data)

# Filter the students DataFrame
filtered_students = students_df[students_df["Grade"] > 90]

# Calculate the average age
average_age = filtered_students["Age"].mean()

# Print the filtered students and average age
print("Filtered Students:")
print(filtered_students)
print("Average Age:", average_age)
````

````python
# Exercise 5: Aggregating Data with Pandas
import pandas as pd
import matplotlib.pyplot as plt

# Load the iris dataset
iris = pd.read_csv("https://gist.githubusercontent.com/netj/8836201/raw/6f9306ad21398ea43cba4f7d537619d0e07d5ae3/iris.csv")
print(iris)
# Check the column names
print(iris.columns)

# Update the column name if necessary
# For example, if the column name is "sepal length (cm)"
sepal_length_min = iris["sepal.length"].min()
sepal_length_max = iris["sepal.length"].max()
sepal_length_mean = iris["sepal.length"].mean()

print("Minimum Sepal Length:", sepal_length_min)
print("Maximum Sepal Length:", sepal_length_max)
print("Mean Sepal Length:", sepal_length_mean)

# Group the data by the "variety" column and calculate the average value for each numeric column
average_variety = iris.groupby("variety").mean()

print("Average Values by variety:")
print(average_variety)

# Create a bar plot showing the average sepal length for each species
average_variety["sepal.length"].plot(kind="bar", title="Average Sepal Length by variety")

# Show the plot
plt.show()

````

````python
# Exercise 6: Data Visualization with Pandas and Matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Load the tips dataset
tips = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv")

# Map "Yes" to 1 and "No" to 0 in the "smoker" column
tips["smoker"] = tips["smoker"].map({"Yes": 1, "No": 0})

# Create a scatter plot using the "total_bill" and "tip" columns, color-coded by the "smoker" column
plt.scatter(tips["total_bill"], tips["tip"], c=tips["smoker"])

# Set plot labels and title
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.title("Scatter Plot of Total Bill vs Tip (Color-coded by Smoker)")

# Show the plot
plt.show()

# Create a box plot to visualize the distribution of the "total_bill" column, grouped by the "day" column
tips.boxplot(column="total_bill", by="day")

# Set plot labels and title
plt.xlabel("Day")
plt.ylabel("Total Bill")
plt.title("Box Plot of Total Bill by Day")

# Show the plot
plt.show()

# Create a bar plot to show the average tip amount for each day of the week
avg_tip_by_day = tips.groupby("day")["tip"].mean()

# Create a bar plot
avg_tip_by_day.plot(kind="bar", rot=0)

# Set plot labels and title
plt.xlabel("Day")
plt.ylabel("Average Tip Amount")
plt.title("Average Tip Amount by Day of the Week")

# Show the plot
plt.show()
````
