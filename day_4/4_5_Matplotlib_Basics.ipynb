# Matplotlib.pyplot Basics

Matplotlib is the baseline data visualization library in Python that provides a wide range of plotting capabilities. It is highly customizable and offers various plotting methods to suit different types of data and visualization requirements.

**Basic Structure of a Matplotlib Plot:**

```python
import matplotlib.pyplot as plt
%matplotlib inline

# Basic Structure
plt.figure(figsize=(8, 6))
plt.<plotype>(,x,y_optional, <plot options> )
plt.title('Plot Title ')
plt.xlabel('x-axis label')
plt.ylabel('y-axis label')
plt.show()

```

- `import matplotlib.pyplot as plt`: access the functions and features provided by the pyplot

- `%matplotlib inline`: this is a magic command used in Jupyter Notebook. It allows the generated plots to be displayed directly within the notebook interface.

- `plt.figure(figsize=(8, 6))`: This creates a new figure object with a specified figure size. The figsize parameter sets the **width and height** of the figure in inches.

- `plt.<plotype>(x, y_optional, <plot options>)`: This is a placeholder for a specific plotting method that you'll choose based on your visualization needs. You will replace <plotype> with the desired plotting method, such as plot, scatter, bar, hist, etc. The x parameter represents the x-axis data, and y_optional represents the y-axis data (if applicable). You can also specify additional options based on the chosen plotting method.

- `plt.title('Plot Title')`: This sets the **title** of the plot.

- `plt.xlabel('x-axis label')`: This sets the label for the x-axis.

- `plt.ylabel('y-axis label')`: This sets the label for the y-axis.

- `plt.show()`: This displays the plot in the notebook. **It is necessary to include this line at the end to visualize the plot**

> **NOTE: you can also save a plot with the savefig() function**

- `plt.savefig('filename.extension')`: This willsave the plot to the specified file in the current working directory or the path you provide. *Matplotlib supports a wide range of file formats, and it determines the appropriate format based on the file extension you specify.*

## Histograms

A histogram is a graphical representation of the distribution of a dataset. It provides a visual summary of the **frequency or count** of values falling within specified intervals, called bins. Histograms are commonly used to understand the underlying distribution and patterns in numerical data.

Use Cases: Histograms are useful in various scenarios, including:

- Understanding Data Distribution: Histograms help visualize and analyze the distribution of continuous or discrete data. They reveal insights about the concentration, spread, and shape of the data.

- Identifying Outliers: Histograms can highlight **extreme or unusual values** that deviate from the main distribution, aiding in outlier detection.


### Interpretation:
By examining a histogram, you can understand the data's central tendency (peaks, mean, median), spread (range, variance), skewness (left or right-leaning distribution), and presence of outliers or clusters.


---


To create a histogram in Python using Matplotlib we use the `.hist()` funciton:

```python
plt.figure(figsize=(8, 6))
plt.hist(data, bins=30, color='steelblue', edgecolor='black')
plt.title('Histogram of Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

- data represents the series (yes single column) data for which we want to create a histogram.
- `bins` determines the number of bins (intervals) to divide the data range into.
- `color` sets the color of the bars in the histogram color options include: 'red', 'blue', 'green', 'orange', 'purple', 'yellow', 'cyan', 'magenta', 'gray' or even hex: '#FF0000' represents red, '#00FF00' represents green, and '#0000FF' represents blue.
- `edgecolor` defines the color of the edges of the bars.

Using this template make a histogram of abnb['number_of_reviews'], change the figure size to 8,2




````python
plt.figure(figsize=(8, 2))
plt.hist(abnb['number_of_reviews'], bins=30, color='steelblue', edgecolor='black')
plt.title('Histogram of Reviews')
plt.xlabel('number_of_reviews')
plt.ylabel('Frequency')
plt.show()
````

## Boxplots

A boxplot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a dataset. It provides a concise summary of the data's central tendency, spread, and skewness.

- **Components**: A boxplot consists of several key components:
  - A box that represents the interquartile range (IQR), which encompasses the middle 50% of the data. The vertical line inside the box denotes the median.
  - Whiskers that extend from the box to the minimum and maximum values within a certain range.
  - Optional outliers, plotted as individual points beyond the whiskers.

- **Data Distribution**: Boxplots offer insights into the shape, spread, skewness, and presence of outliers in a dataset. They are particularly useful for comparing multiple distributions side by side.

- **Visualization**: Boxplots provide a clear visual representation of summary statistics and identify potential outliers or skewness. They offer a compact way to display distribution characteristics without displaying the entire dataset.

- **Use Cases**: Boxplots are useful in various scenarios, including:

  - **Comparing Distributions**: Boxplots are effective for comparing multiple distributions within one plot. They enable quick comparisons of central tendency, spread, and skewness across different groups or categories.

  - **Identifying Outliers**: Boxplots visually highlight potential outliers in the dataset, helping to identify extreme values that deviate significantly from the bulk of the data.

  - **Assessing Skewness**: Boxplots can reveal the skewness of a distribution by comparing the lengths of the whiskers. Unequal whisker lengths indicate skewness towards the lower or upper values.

- **Interpretation**: By examining a boxplot, you can understand the data's central tendency (median), spread (IQR), presence of outliers, and skewness (asymmetric whisker lengths).

Boxplots are widely used for exploratory data analysis, providing a concise summary of the distribution of data and facilitating comparisons between multiple groups. They help identify outliers, assess skewness, and gain insights into the data's central tendency and spread.

Basic Boxplot Template

To create a boxplot in Python using Matplotlib we use the `.boxplot()` funciton:

```python

plt.figure(figsize=(8, 6))
plt.boxplot(series, notch=True, patch_artist=True)
plt.title('Boxplot of Single Column')
plt.xlabel('Series')
plt.ylabel('Value')
plt.show()
```

>Comparing 2 or more Series requires a little setup, specifically creating a list of series objects then optionally providing a label for each series.

```python
plt.figure(figsize=(8, 6))

plt.boxplot(list_of_series, labels=['Dataset 1', 'Dataset 2', 'Dataset 3'], notch=True, patch_artist=True)
plt.title('Boxplot of Datasets')
plt.xlabel('Dataset')
plt.ylabel('Value')
plt.show()

```
- list_of_series ilist of series data to plot
- labels is a list of labels for each item in the series list
- notch=True creates a notched boxplot, which displays a confidence interval around the median.
- patch_artist=True fills the boxes with color.

## Bar Charts

A bar chart, also known as a bar graph, is a visualization that presents categorical data using rectangular bars. It is commonly used to compare and display the values of different categories or groups.

- **X-axis and Y-axis**: In a bar chart, the x-axis represents the categories or groups, while the y-axis represents the corresponding values or frequencies.

- **Bar Length**: The length or height of each bar represents the magnitude or value associated with the category or group it represents. The bars can be vertical (column chart) or horizontal (bar chart).

- **Categorical Data**: Bar charts are particularly useful when visualizing categorical or qualitative data. Each category is represented by a separate bar, allowing easy comparison between different categories.


- **Use Cases**: Bar charts are useful in various scenarios, including:

  - **Comparison**: Bar charts allow you to compare the values of different categories or groups. They help identify the highest or lowest values, spot trends, and make comparisons between related data points.

  - **Distribution**: Bar charts can be used to visualize the distribution or frequency of categorical data. They show the count or proportion of each category, revealing patterns and identifying dominant categories.

  - **Ranking**: Bar charts are effective for ranking or ordering categories based on their values. They visually represent the differences in magnitude or quantity between categories.

## Introduction to `plot()` Function in Matplotlib

The `plot()` function in Matplotlib is used to create line plots and scatter plots. It allows you to visualize data points and patterns by connecting them with lines or markers.

- **Data Points**: The `plot()` function accepts input data points as arguments for the x and y axes. These data points can be provided as arrays, lists, or other iterable objects.

- **Line Plot**: By default, the `plot()` function creates a line plot, where the data points are connected by straight lines. This is useful for visualizing trends, changes over time, or continuous data.

- **Markers**: You can add markers to the data points using the `marker` parameter in the `plot()` function. Markers can represent individual data points and help distinguish between them.

- **Line Styles**: The `plot()` function allows you to specify different line styles using the `linestyle` parameter. You can choose between solid lines (`'-'`), dashed lines (`'--'`), dotted lines (`':'`), and more.

- **Colors**: You can customize the color of the lines and markers using the `color` parameter. Matplotlib provides a wide range of color options, such as named colors (e.g., `'red'`, `'blue'`), hexadecimal color codes (e.g., `'#FF0000'`, `'#0000FF'`), and more.

- **Visualization**: The `plot()` function creates a plot object that can be further customized by adding labels, titles, legends, and more using additional functions provided by Matplotlib.

- **Use Cases**: The `plot()` function is commonly used for visualizing trends, patterns, and relationships in data. It is suitable for line plots, scatter plots, and other types of plots where data points are connected or marked individually.

```python
plt.figure(figsize=(8, 6))
plt.plot(x, y, marker='o', linestyle='-', color='steelblue')
plt.title('Plot Title')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.legend(['Legend'])
plt.show()
```


# Fix the X-Axis

When working with dates on the x-axis of your plots, the `matplotlib.dates` module aka `mdates` provides tools for formatting and customizing the appearance of the dates.

The DateFormatter class allows you to specify the format of dates using various format codes to represent the year, month, day, hour, minute, and more. For example, %b %Y represents the abbreviated month followed by the four-digit year.

- `%Y`: Four-digit year (e.g., 2023)
- `%y`: Two-digit year (e.g., 23)
- `%m`: Month as a zero-padded decimal number (e.g., 01 for January)
- `%b`: Abbreviated month name (e.g., Jan)
- `%B`: Full month name (e.g., January)
- `%d`: Day of the month as a zero-padded decimal number (e.g., 01)
- `%w`: Weekday as a decimal number, where Sunday is 0 and Saturday is 6
- `%W`: Week number of the year as a decimal number, with Monday as the first day of the week (e.g., 00 to 53)



> the MonthLocator class comes in handyhelps determine the locations where the major ticks will be placed, such as at the beginning of each month. This ensures that the x-axis displays a **single date for each month**, enhancing readability.

The `plt.gca().xaxis.set_major_locator()` and `plt.gca().xaxis.set_major_formatter()` functions in Matplotlib provide convenient ways to customize the major tick positions and formatting of the x-axis.

- `set_major_locator()` is used to specify where the major tick marks should appear on the x-axis.

- `set_major_formatter()` is used to define how the major tick labels should be formatted on the x-axis.

To make use of these functions, you can call `plt.gca().xaxis.set_major_locator()` to set the desired tick positions on the x-axis, and then use `plt.gca().xaxis.set_major_formatter()` to format the tick labels accordingly.

> gca() stands for get current axis



````python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

result["Date"] = pd.to_datetime(result["Date"])

plt.figure(figsize=(8, 6))
plt.plot(result["Date"], result["High"], marker='o', linestyle='-', color='steelblue')
plt.title('Lineplot of MSFT high price')
plt.xlabel('Date')
plt.ylabel('High')

# Format x-axis to show only the month
months = mdates.MonthLocator()
plt.gca().xaxis.set_major_locator(months)
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))

plt.xticks(rotation=45)
plt.show()
````
